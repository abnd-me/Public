**1. Reading Dataset**

```python
import pandas as pd
#Reading dataset
df1 = pd.read_csv('/content/titanic_dataset.csv')

df1
```

**2. Make PassengerId as index**

```python
#set PassengerId as index
df1.set_index('PassengerId' , inplace=True)

df1.head(10)
```

**3. Display Basic information about dataset**

```python

df1.info()
```

**4. Manage Null values**

```python
df2 = df1
#Filling null values in numerical columns using mean
df3 = df2.fillna(df2.mean())

df3.info()
```

```python
#Removing null values of catagorical columns

df3['Embarked'].fillna(df3['Embarked'].mode()[0], inplace=True) #using mode


constant_value = 0
df3['Cabin'].fillna(constant_value, inplace=True )  #using constant value


df3.info()
```

**5. Handle Outliers**

```python
#Outlies are handled using z score method
# Three nnumerical columns Age SibSp Fare are chosen to handle outliers
outlier_column = ['Age' , 'SibSp' , 'Fare']

df4 = df3

#Finding mean and standard deviation of selected columns
for column in outlier_column:
  data_mean = df4[column].mean()
  data_std = df4[column].std()

  z_score = (df4[column] - data_mean)/data_std #calculating z score

  threshold = 2                                #defining threshold value

  df4[column] = df4[column].where(abs(z_score) < threshold) #Handling outliers

print(df4)
```

**6. MinMax Scaling**

```python
from sklearn.preprocessing import MinMaxScaler

df = df4

#Choosing Sruvived as traget
target = df['Survived']

#Extract the features (excluding 'Survived' and  non-numeric columns)
features = df.drop(['Survived', 'Name', 'Sex', 'Ticket', 'Cabin', 'Embarked'], axis=1)

# Creating the MinMaxScaler
scaler = MinMaxScaler()

# Fit the scaler
scaler.fit(features)

# applying the normalization transformation feature set using transform
scaled_features = scaler.transform(features)

# Creating a new DataFrame with the scaled features
df_scaled = pd.DataFrame(scaled_features, columns=features.columns, index=features.index)

# Add the target variable(Survived) back to the DataFrame
df_scaled['Survived'] = target

# Display the new DataFrame after scaling
print(df_scaled)
```
