### we have to extract loan types and assign them to new column for training

```python
# extract individual loan types
def extract_loan_types(loan_str):
    loan_types = [s.strip() for s in loan_str.split(',')]

    flat_loan_types = []
    for loan_type in loan_types:
        if 'and' in loan_type:
            flat_loan_types.extend(loan_type.split('and'))
        else:
            flat_loan_types.append(loan_type)


    return list(set([loan.strip() for loan in flat_loan_types]))

df['Individual_Loan_Types'] = df['Type_of_Loan'].apply(extract_loan_types)

print(df[['Type_of_Loan', 'Individual_Loan_Types']].head())
```

```python
# Get unique individual loan types
all_loan_types = set()
for loan_types_list in df['Individual_Loan_Types']:
    all_loan_types.update(loan_types_list)

for loan_type in all_loan_types:
    df[loan_type] = df['Individual_Loan_Types'].apply(lambda x: 1 if loan_type in x else 0)

df.drop(columns=['Individual_Loan_Types'], inplace=True)
print(df.head())
```

```python
from sklearn.model_selection import train_test_split

X = df_encoded_.drop(['ID', 'Month', 'Customer_ID', 'Name',  'SSN','Type_of_Loan','Credit_Score'], axis=1)  # Features
y = df_encoded_['Credit_Score']  # Target variable
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

X_val, X_test1, y_val, y_test1 = train_test_split(X_test, y_test, test_size=0.5, random_state=42)
```

#Model Training

## Hyperparameter Tuning

```python
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score


# Define models
models = {
    'Logistic Regression': LogisticRegression(),
    'Decision Tree': DecisionTreeClassifier(),
    'Random Forest': RandomForestClassifier(),
    'K-Nearest Neighbors': KNeighborsClassifier()
}

# Define hyperparameter grids
param_grids = {
    'Logistic Regression': {'C': [0.001, 0.01, 0.1, 1, 10, 100]},
    'Decision Tree': {'max_depth': [None, 10, 20, 30], 'min_samples_split': [2, 5, 10]},
    'Random Forest': {'n_estimators': [100, 200, 300], 'max_depth': [None, 10, 20, 30], 'min_samples_split': [2, 5, 10]},
    'K-Nearest Neighbors': {'n_neighbors': [3, 5, 7, 9]}
}

# Hyperparameter tuning and model training
best_models = {}
for model_name, model in models.items():
    grid_search = GridSearchCV(model, param_grid=param_grids[model_name], cv=5, scoring='accuracy')
    grid_search.fit(X_train, y_train)
    best_models[model_name] = grid_search.best_estimator_

# Evaluate models
for model_name, model in best_models.items():
    y_pred = model.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)
    print(f"{model_name}: Accuracy = {accuracy}")
```

```python

from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

param_grid2 = {
    'n_estimators': [100, 200, 300 ,],
    'max_depth': [None, 10, 20, ],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

# Instantiate Random Forest classifier
rf_classifier = RandomForestClassifier(random_state=42)

# Instantiate GridSearchCV
grid_search = GridSearchCV(estimator=rf_classifier, param_grid=param_grid2, cv=5, scoring='accuracy')

# Perform grid search to find the best hyperparameters
grid_search.fit(X_train, y_train)

# Find parameters and best score
best_params = grid_search.best_params_
best_score = grid_search.best_score_

print("Best Parameters:", best_params)
print("Best Score:", best_score)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report

rfo = RandomForestClassifier(random_state=42 ,n_estimators = 300 , max_depth = None , min_samples_split = 5 , min_samples_leaf = 1)
rfo.fit(X_train , y_train)

#predictions
y_pred_rfo = rfo.predict(X_test)

# classification report
print("Classification Report:")
print(classification_report(y_test, y_pred_rfo))

#  accuracy score
accuracy = accuracy_score(y_test, y_pred_rfo)
print("Accuracy Score:", accuracy)
```
