```python
import pandas as pd

df = pd.read_csv('/content/Students_Performance.csv')

df.head(10)
```

## A. males and females participated in the test

```python
gender = df['gender'].value_counts()
print(gender)
```

### B.Student's parental level of education

```python
education = df['parental level of education'].value_counts()
print(education)
```

##C. Who scores the most on average for math, reading and writing

```python
average_scores_gender = df.groupby('gender').mean()[['math score', 'reading score', 'writing score']]
average_scores_course = df.groupby('test preparation course').mean()[['math score', 'reading score', 'writing score']]

print("Average Scores based on Gender:\n", average_scores_gender)
print("\nAverage Scores based on Test Preparation Course:\n", average_scores_course)
```

## D. What do you think about the scoring variation for math, reading and writing

```python
score_variation_gender = df.groupby('gender').std()[['math score', 'reading score', 'writing score']]
score_variation_course = df.groupby('test preparation course').std()[['math score', 'reading score', 'writing score']]

print("Score Variation based on Gender:\n", score_variation_gender)
print("\nScore Variation based on Test Preparation Course:\n", score_variation_course)
```

## E. Give bonus points to the top 25% of students based on their math score,


```python

threshold = df['math score'].quantile(0.75)

top_25 = df[df['math score'] >= threshold] #students with math scores in the top 25%

#add bonus point
bonus_points = 10
top_25['math score'] += bonus_points

print(top_25)
```

#2 Case Study on Testing of Hypothesis


```python
df1 = pd.read_csv('/content/Sales_add.csv')

df1.head(10)
```

##Check whether there is any  increase in sales after stepping into digital marketing

```python
import scipy.stats as stats

t_stat, p_value = stats.ttest_rel(df1['Sales_before_digital_add(in $)'], df1['Sales_After_digital_add(in $)'])

# Check the p-value
print("p-value:", p_value)

# Compare p-value with the significance level
if p_value < 0.05:
    print("There is a significant increase in sales after digital marketing.")
else:
    print("There is no significant increase in sales after digital marketing.")
```

##Check whether there is any dependency between the features “Region” and “Manager”.


```python

contingency_table = pd.crosstab(df1['Region'], df1['Manager'])

# Perform the Chi-square test
chi2_stat, p_value, dof, expected = stats.chi2_contingency(contingency_table)

# Check the p-value
print("p-value:", p_value)

# Compare p-value with the significance level
if p_value < 0.05:
    print("There is a significant dependency between 'Region' and 'Manager'.")
else:
    print("There is no significant dependency between 'Region' and 'Manager'.")
```
